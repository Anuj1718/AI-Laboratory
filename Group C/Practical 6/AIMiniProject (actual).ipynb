```python
import gradio as gr

class Employee:
    def __init__(self, name, role, experience, performance_scores):
        self.name = name
        self.role = role
        self.experience = experience
        self.performance_scores = performance_scores

    def evaluate_performance(self):
        overall_score = sum(self.performance_scores.values()) / len(self.performance_scores)

        if self.experience > 5 and overall_score > 4.5:
            return "üåü Excellent Performer"
        elif self.experience > 2 and overall_score > 3.5:
            return "üëç Good Performer"
        elif self.experience <= 2 and overall_score > 2.5:
            return "üëå Developing Performer"
        else:
            return "‚ö†Ô∏è Needs Improvement"

def evaluate_performance(name, role, experience, task_completion, collaboration, punctuality, initiative, feedback):
    score = 0

    # Task Completion mapped to 1‚Äì5 scale
    if task_completion >= 90:
        score += 5
    elif task_completion >= 75:
        score += 4
    elif task_completion >= 60:
        score += 3
    elif task_completion >= 45:
        score += 2
    else:
        score += 1

    # Sum other categories (1‚Äì5)
    score += collaboration
    score += punctuality
    score += initiative
    score += feedback

    # Average out
    avg_score = score / 6

    # Create employee object
    performance_scores = {
        "Task Completion": task_completion,
        "Collaboration": collaboration,
        "Punctuality": punctuality,
        "Initiative": initiative,
        "Manager Feedback": feedback
    }

    emp = Employee(name, role, experience, performance_scores)
    evaluation = emp.evaluate_performance()

    # Summary messages
    if "Excellent" in evaluation:
        summary = "Outstanding performance across all metrics. Keep up the great work!"
    elif "Good" in evaluation:
        summary = "The employee is performing well in most areas, with minor room for improvement."
    elif "Developing" in evaluation:
        summary = "Performance is acceptable but needs improvement in some areas."
    else:
        summary = "Several areas require immediate improvement. Training and mentoring advised."

    report = (
        f"üë§ **Name**: {name}\n"
        f"üíº **Role**: {role}\n"
        f"üìÖ **Experience**: {experience} years\n\n"
        f"üìä **Scores**:\n"
        f"‚Ä¢ Task Completion: {task_completion}%\n"
        f"‚Ä¢ Team Collaboration: {collaboration}/5\n"
        f"‚Ä¢ Punctuality: {punctuality}/5\n"
        f"‚Ä¢ Initiative: {initiative}/5\n"
        f"‚Ä¢ Manager Feedback: {feedback}/5\n\n"
        f"üîé **Average Score**: {round(avg_score, 2)} / 5\n"
        f"‚úÖ **Evaluation**: {evaluation}\n\n"
        f"üí¨ **Summary**: {summary}"
    )

    return report

# Input components
inputs = [
    gr.Textbox(label="Employee Name"),
    gr.Textbox(label="Role / Designation"),
    gr.Slider(0, 40, step=1, label="Years of Experience"),
    gr.Slider(0, 100, label="Task Completion Rate (%)"),
    gr.Slider(1, 5, step=1, label="Team Collaboration"),
    gr.Slider(1, 5, step=1, label="Punctuality"),
    gr.Slider(1, 5, step=1, label="Initiative"),
    gr.Slider(1, 5, step=1, label="Manager Feedback")
]

output = gr.Textbox(label="üìÑ Performance Evaluation Report")

# Gradio app
app = gr.Interface(
    fn=evaluate_performance,
    inputs=inputs,
    outputs=output,
    live=True,
    title="üß† Expert System for Employee Performance Evaluation",
    description="Enter employee details and performance metrics to get a real-time AI-driven evaluation report."
)

app.launch(share=True)
```
